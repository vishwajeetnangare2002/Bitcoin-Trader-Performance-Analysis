```python
fear_df = fear_df.drop_duplicates()

trade_df = trade_df.drop_duplicates()

print("Duplicates Removed")
```

```python
# Convert 'date' column to datetime format
fear_df['date'] = pd.to_datetime(fear_df['date'])

# Check the data type
fear_df['date'].dtype
```

```python
trade_df['Timestamp IST'] = pd.to_datetime(
    trade_df['Timestamp IST'],
    format='%d-%m-%Y %H:%M'
)
```

```python
trade_df['Date'] = trade_df['Timestamp IST'].dt.normalize()
```

```python
fear_df['date'] = pd.to_datetime(fear_df['date'])
```

```python
trade_df[['Timestamp','Date']].head()
```

```python
fear_df.head()
```

```python
merged_df = pd.merge(
    trade_df,
    fear_df[['date','classification']],
    left_on='Date',
    right_on='date',
    how='left'
)
```

```python
merged_df.head()
```

```python
print("Shape of merged dataset:")
print(merged_df.shape)

print("\nColumn Names:")
print(merged_df.columns)
```

```python
merged_df.describe()
```

```python
merged_df['classification'].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='classification',
    data=merged_df,
    order=merged_df['classification'].value_counts().index
)

plt.title("Market Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Trades")
plt.xticks(rotation=20)

plt.show()
```

```python
print("Total Profit/Loss")

print(merged_df['Closed PnL'].sum())
```

```python
print("Average Profit/Loss")

print(merged_df['Closed PnL'].mean())
```

```python
print("Highest Profit")

print(merged_df['Closed PnL'].max())
```

```python
print("Largest Loss")

print(merged_df['Closed PnL'].min())
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    merged_df['Closed PnL'],
    bins=50,
    kde=True
)

plt.title("Distribution of Closed PnL")

plt.show()
```

```python
profit_by_sentiment = merged_df.groupby('classification')['Closed PnL'].mean()

print(profit_by_sentiment)
```

```python
plt.figure(figsize=(8,5))

profit_by_sentiment.plot(kind='bar')

plt.title("Average Closed PnL by Market Sentiment")

plt.ylabel("Average Closed PnL")

plt.show()
```

```python
merged_df['Side'].value_counts()
```

```python
plt.figure(figsize=(6,6))

merged_df['Side'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel("")

plt.title("Buy vs Sell Trades")

plt.show()
```

ADVANCE EDA

```python
coin_profit = merged_df.groupby('Coin')['Closed PnL'].sum().sort_values(ascending=False)

coin_profit.head(10)
```

```python
plt.figure(figsize=(12,6))

coin_profit.head(10).plot(kind='bar')

plt.title("Top 10 Most Profitable Coins")
plt.xlabel("Coin")
plt.ylabel("Total Closed PnL")

plt.show()
```

```python
coin_loss = merged_df.groupby('Coin')['Closed PnL'].sum().sort_values()

coin_loss.head(10)
```

```python
plt.figure(figsize=(12,6))

coin_loss.head(10).plot(kind='bar', color='red')

plt.title("Top 10 Loss Making Coins")

plt.show()
```

```python
avg_coin_profit = merged_df.groupby('Coin')['Closed PnL'].mean().sort_values(ascending=False)

avg_coin_profit.head(10)
```

```python
sentiment_profit = merged_df.groupby('classification')['Closed PnL'].agg(['count','mean','sum'])

sentiment_profit
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=merged_df,
    x='classification',
    y='Closed PnL',
    estimator=np.mean
)

plt.title("Average Profit by Market Sentiment")

plt.show()
```

```python
merged_df['Trade Result'] = np.where(
    merged_df['Closed PnL'] > 0,
    'Profit',
    'Loss'
)
```

```python
merged_df['Trade Result'].value_counts()
```

```python
plt.figure(figsize=(6,6))

merged_df['Trade Result'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Winning vs Losing Trades")

plt.ylabel("")

plt.show()
```

```python
pd.crosstab(
    merged_df['classification'],
    merged_df['Trade Result']
)
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    pd.crosstab(
        merged_df['classification'],
        merged_df['Trade Result']
    ),
    annot=True,
    cmap='Blues',
    fmt='d'
)

plt.title("Win Rate by Market Sentiment")

plt.show()
```

```python
top_traders = merged_df.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False)

top_traders.head(10)
```

```python
plt.figure(figsize=(12,5))

top_traders.head(10).plot(kind='bar')

plt.title("Top 10 Profitable Traders")

plt.ylabel("Profit")

plt.show()
```

```python
daily_trades = merged_df.groupby('Date').size()

daily_trades.head()
```

```python
plt.figure(figsize=(15,5))

daily_trades.plot()

plt.title("Daily Number of Trades")

plt.ylabel("Trades")

plt.show()
```
